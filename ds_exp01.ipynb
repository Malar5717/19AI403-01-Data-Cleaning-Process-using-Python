```python
df.dropna(axis=0)
# drop null values, along row
```

```python
df.fillna(0)
# fill all null values, with 0
```

```python
df.fillna(method='bfill')
# Backward fill: fills missing values with the next valid value below
```

```python
df.fillna(method='ffill')
# Forward fill: fills missing values with the previous valid value above.
```

```python
df['watchers'].fillna(value=df['watchers'].mean())
# replaces null values in 'watchers' coloumn with the coloumns mean
```

```python
df['watchers'].fillna(value=df['watchers'].mean(),inplace=True)
df
```

Remove Outliers using IQR

```python
Q1 = df['watchers'].quantile(0.25)
Q3 = df['watchers'].quantile(0.75)
IQR = Q3 - Q1

df_clean = df[(df['watchers'] >= Q1 - 1.5 * IQR) & (df['watchers'] <= Q3 + 1.5 * IQR)]

# IQR -> Interquartile Range.
# tells us the middle 50% of the data (difference between the 75th and 25th percentiles of the data)
```

```python
df_clean.head()
```

```python
df_clean.tail()
```

```python
from scipy.stats import zscore

numeric_cols = ['watchers', 'rating', 'num_episodes']
z_scores = zscore(df[numeric_cols])

# Z-score tells you how far a value is from the mean, in terms of standard deviations
# If the absolute Z-score > 3, it is usually considered an outlier

mask = (abs(z_scores) < 3).all(axis=1)
df_cleane = df[mask]
```
